```python
from bs4 import BeautifulSoup
import csv
import requests
```

```python
# Muscbrainz URL für Deutschland unter dem Reiter Künstler
url = "https://musicbrainz.org/area/85752fda-13c4-31a3-bee5-0e5cb1f51dad/artists"

req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')

# Künstler sind im folgenden URL Muster eingebettet - 'a[title=":edward"] bdi'
print(soup.prettify())
```

```python
# Muscbrainz URL für Deutschland unter Berücksichtigung der pages
base_url = "https://musicbrainz.org/area/85752fda-13c4-31a3-bee5-0e5cb1f51dad/artists?page="

# Liste für alle Künstler Namen
all_artist_names = []

# Iteriere über die Anzahl Pages - Vorgegeben durch die Range
for page_number in range(1, 911):  # Alle Pages scrapen (insgesamt gibt es 910)
    # Url für die aktuelle Page
    url = base_url + str(page_number)

    # GET request für die jeweilige URL
    req = requests.get(url)
    soup = BeautifulSoup(req.content, 'html.parser')
    
    # Suche nach den Künstler Namen in dem vorgegebenen Muster
    artist_names = [artist.text for artist in soup.select('a[title] bdi')]
    
    # Füge den Künstlernamen zu der Liste "all_artist_names" hinzu
    all_artist_names.extend(artist_names)

# Drucke jeden Namen in der Künstler Liste
for artist_name in all_artist_names:
    print(artist_name)
```

```python
# Liste in CSV speichern

filename = "musicbrainz_country_de.csv"
data = all_artist_names

with open(filename, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    for value in data:
        writer.writerow([value])
```
