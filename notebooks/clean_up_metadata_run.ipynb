```python
import pandas as pd
import spacy

nlp = spacy.load('de_core_news_sm')

def is_word(token):
    return not token.is_stop and not token.is_punct and not token.is_space

def count_words(text):
    doc = nlp(text)
    return len([token for token in doc if is_word(token)])

df['word_count'] = df['lyrics'].apply(lambda x: count_words(x))

print(df)
```

```python
fill_value = 0
df['release_year'] = df['release_year'].fillna(fill_value)
df['release_year'] = df['release_year'].astype(int)

df['decade'] = df['decade'].fillna(fill_value)
df['decade'] = df['decade'].astype(int)
```
