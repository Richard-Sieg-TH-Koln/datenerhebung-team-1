```python
import pandas as pd
import json

with open('lyrics_stadtaffe.json') as f:
    data = json.load(f)

tracks = data['tracks']

titles = []
artists = []
lyrics = []

for track in tracks:
    song = track['song']
    titles.append(song['title'])
    artists.append(song['artist'])
    lyrics.append(song['lyrics'])

df = pd.DataFrame({'Titel': titles, 'Künstler': artists, 'Text': lyrics})
print(df)
```

```python
df.to_csv('lyrics_stadtaffe.csv', index=False)
```
