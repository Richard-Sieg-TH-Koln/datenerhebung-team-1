```python
import pandas as pd
import spacy

nlp = spacy.load('de_core_news_sm')

def is_word(token):
    return not token.is_stop and not token.is_punct and not token.is_space

def count_words(text):
    doc = nlp(text)
    return len([token for token in doc if is_word(token)])

df = pd.read_csv('dataset_cleaned_file_v4.csv')

df['word_count'] = df['lyrics'].apply(lambda x: count_words(x))

print(df)

```

```python
word_counts = df['word_count']
print(word_counts)
```

```python
df.to_csv('./dataset_cleaned_file_v4_edited.csv', index=False)
```
