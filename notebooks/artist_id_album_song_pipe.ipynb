```python
def test(artist_name):
    # List to save the data -> Later transformed to dataframe
    data = []
    # Search artist from test set -> Find primary_artist_id
    artist = genius.search(artist_name)
    if artist is not None:
        # Get all primary_artist ids
        artist_ids = [hit['result']['primary_artist']['id'] for hit in artist['hits']]
        # Filter only for unique artist and artist_id
        artist_ids_unique = set(artist_ids)
        
        for id in artist_ids_unique:
            # Search artist_id -> Find albums
            albums = genius.artist_albums(id)
            if albums is not None:
                # Create list of lists containing artist name, ID, album name, and release date
                albums_data = [[artist_name, id, alb['name'], alb['release_date_for_display']] for alb in albums['albums']]
                # Append albums_data to the data
                data.extend(albums_data)
                print(albums_data)

            else:
                pass
        
        return data

    else:
        pass


# Create empty DataFrame to store the data from the data list
df = pd.DataFrame(columns=["Artist", "Artist_ID", "Album", "Release_Date"])

# Input artist name
for name in artist_test_df["Interpreten"]:
    artist_name = name

    # Call function test
    albums_data = test(artist_name)
    if albums_data is not None:
        # Store the data in the DataFrame
        df = df.append(pd.DataFrame(albums_data, columns=["Artist", "Artist_ID", "Album", "Release_Date"]), ignore_index=True)
```
