# Create a dataframe containing Artist, Album and Songs

```python
# Search for the "Artist Name" and retrieve all "Primary Artist IDs" from Genius
# With the "Primary Artist IDs" search for the "Album" of the "Artist" and retrieve the "Album Name", "Album ID" and "Album Release Date"

def artist_album_ids(artist_name):
    # List to save the data -> Later transformed to dataframe
    data = []
    # Search artist from test set -> Find primary_artist_id
    artist = genius.search(artist_name)
    if artist is not None:
        # Get all primary_artist ids
        artist_ids = [hit['result']['primary_artist']['id'] for hit in artist['hits']]
        # Filter only for unique artist and artist_id
        artist_ids_unique = set(artist_ids)
        
        for artist_id in artist_ids_unique:
            # Search artist_id -> Find albums
            albums = genius.artist_albums(artist_id)
            if albums is not None:
                # Create list of lists containing artist name, ID, album name, and release date
                albums_data = [[artist_name, artist_id, alb['name'], alb['id'],alb['release_date_for_display']] for alb in albums['albums']]
                # Append albums_data to the data
                data.extend(albums_data)

            else:
                pass
        
        return data

    else:
        pass


# Create empty DataFrame to store the data from the data list
df_albums = pd.DataFrame(columns=["Artist", "Artist_ID", "Album", "Album_ID", "Release_Date"])

# Input artist name
for name in tqdm(artist_df["Artist_Name"]):
    artist_name = name

    # Call function test
    albums_data = artist_album_ids(artist_name)
    if albums_data is not None:
        # Store the data in the DataFrame
        df_albums = df_albums.append(pd.DataFrame(albums_data, columns=["Artist", "Artist_ID", "Album", "Album_ID", "Release_Date"]), ignore_index=True)
```

```python
# Search for the "Artist Name" and iterate through the Genre Tags until one matches the Genres from the List

def lastfm_artist_genre(artist_name):
    
    # Set the parameters for the API request
    params = {
        "method": "artist.getTopTags",
        "artist": artist_name,
        "api_key": API_KEY,
        "format": "json"
    }
    
    # Send GET request to the API endpoint with the specified parameters
    response = requests.get(ENDPOINT_URL, params=params)
    
    # Extract the JSON data from the response
    data = response.json()
    
    try:
        # Get the list of tags for the artist's genre from the JSON data
        tags_genre = data["toptags"]["tag"]
        
        # Check if the list of tags is not empty
        if len(tags_genre) != '':
            # Iterate over each tag in the list
            for tag in tags_genre:
                # Convert the tag name to lowercase
                top_tag = tag["name"].lower()
                
                # Check if the lowercase tag is in the genre_list
                if top_tag in genre_list:
                    # Return the lowercase tag as the genre
                    return top_tag
        else:
            # Return None if the list of tags is empty
            return None
    
    except (KeyError, IndexError):
        # Return an empty list if there is a KeyError or IndexError
        return []

# Define a list of genres
genre_list = ['ambient', 'metal', 'deutsch rap', 'deutschrap', 'german rap', 'deutschrock', 'dubstep', 
              'electro', 'emo', 'gothic', 'hip hop', 'hip-hop', 'indie', 'instrumental', 'ndw', 'party',
              'pop', 'pop rock', 'pornorap', 'post-rock', 'punk rock', 'rap', 'reggae', 
              'rock', 'schlager', 'swing', 'vocal', 'cloud rap', 'alternative rap', 'new wave', 'synth pop', 'karneval']

# Create an empty dataframe to store artist genres
artist_genre_df = pd.DataFrame(columns=["artist", "artist_id", "genre"])

# Iterate over each row in the df_artist_unique dataframe
for index, row in df_artist_unique.iterrows():
    # Get the artist name and artist ID from the current row
    artist_name = row['Artist']
    artist_id = row['Artist_ID']

    # Retrieve the genre for the artist
    genre = lastfm_artist_genre(artist_name)

    # Append the artist, artist ID, and genre to the artist_genre_df dataframe
    artist_genre_df = artist_genre_df.append({"artist": artist_name, "artist_id": artist_id, "genre": genre}, ignore_index=True)
```

```python
# Define a dictionary for mapping genres to categories
genre_mapping = {
    'Pop': ['ambient', 'pop', 'pop rock', 'synth pop', 'ndw', 'party', 'swing', 'vocal', 'new wave'],
    'Schlager': ['schlager', 'karneval'],
    'Rap': ['deutsch rap', 'deutschrap', 'german rap', 'pornorap', 'rap', 'alternative rap', 'cloud rap', 'hip hop', 'hip-hop', 'reggae'],
    'Rock': ['metal', 'deutschrock', 'punk rock', 'post-rock', 'rock', 'emo', 'gothic', 'indie']
}

# Function to map a genre to a category based on the genre_mapping dictionary
def map_genre(value):
    for genre, categories in genre_mapping.items():
        if value in categories:
            return genre
    return None

# Create a new column 'genre_cat' in the artist_genre_df dataframe
# Apply the map_genre function to the 'genre' column to map genres to categories
artist_genre_df['genre_cat'] = artist_genre_df['genre'].apply(map_genre)
```
