```python
def search_lyrics_by_artist(artist_name):
    # Search artist from musicbrainz_country_de
    # Search for the latest 5 released songs
    artist = genius.search_artist(artist_name, sort="release_date", max_songs=30)
    if artist is not None:
        # List to save the data -> Later transformed to dataframe
        data = []
        # Generate a list of song titles for each artist
        titles = [song.title for song in artist.songs]
        for title in titles:            
            # Search for specific song by title and artist
            song = genius.search_song(title, artist_name)
            if song is not None:
                # Append data to the list for each song
                data.append([artist_name,
                            title,
                            song._body['release_date_for_display'],
                            song._body['language'],
                            song._body['stats'],
                            song._body['release_date_components'],
                            song.full_title,
                            song.id,
                            song.lyrics])
            else:
                pass

        return data
    
    else:
        pass


# Create empty DataFrame to store the data from the data list
df = pd.DataFrame(columns=["artist", "song", "release_date_for_display", "language","stats","release_date_components","full_title", "song_id", "lyrics"])


# Input artist name
for name in artist_test_df["artist"]:
    artist_name = name

    # Call function search_lyrics_by_artist
    lyrics_data = search_lyrics_by_artist(artist_name)
    if lyrics_data is not None:
        fieldnames = ["artist", "song", "release_date_for_display", "language","stats","release_date_components","full_title", "song_id", "lyrics"]
        df = pd.DataFrame(lyrics_data, columns=fieldnames)

        # Write the DataFrame to a CSV file
        df.to_csv('metadata_run.csv', mode='a', index=False, header=not os.path.exists('metadata_run.csv'))
```
